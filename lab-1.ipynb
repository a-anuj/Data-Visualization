```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,4,12,8,10]

a = [6,7,8,9,4,17]
b = [1,10,2,12,3,13]

plt.scatter(x,y,marker="o",color="green")
plt.plot(x,y,linestyle=':')

plt.scatter(a,b,marker="x",color="green")
plt.plot(a,b,linestyle='-')

plt.title("Sample Data Vizualization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
