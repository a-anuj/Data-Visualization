```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
df = pd.read_csv("practiceCSV.csv")
print(df)
```

```python
plt.bar(df["Department"],df["Salary"],width=0.3)    
```

```python
grouped_value = df.groupby("Department")["Salary"].mean()
grouped_value.plot(kind="bar")
```

```python
plt.style.use("seaborn-v0_8-darkgrid")
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,"g-",label="y1 = sin(x)")
plt.plot(x,y2,"b-",label="y2 = cos(x)")
plt.legend(loc="upper right",bbox_to_anchor=(1.3, 1.0))
plt.show()
```
